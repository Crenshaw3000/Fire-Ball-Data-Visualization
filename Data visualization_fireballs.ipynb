```python
import pandas as pd
import numpy as np
import datetime
```

```python
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('cneos_fireball_data.csv')
df
```

```python
df.dtypes
```

```python
df.describe()
```

```python
df.describe(include = 'object')
```

```python
#Take the log of Caculated Total Impact Energry
df['Log Impact Energy']= np.log(df['Calculated Total Impact Energy (kt)'])
df
```

```python
#Parse year out of Peak Brightness Date/Time colument to make a Year column
df['Year'] = pd.to_datetime(df['Peak Brightness Date/Time (UT)']).dt.year
df
```

```python
#View counts of the Years
df['Year'].value_counts()
```

```python
#add Counts to dataframe and labeled the column Events_by_year
df['Year'] = pd.to_datetime(df['Peak Brightness Date/Time (UT)']).dt.year
df['Events_by_year'] = df.groupby('Year')['Year'].transform('count')
df
```

```python
df.describe()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df.plot(x='Year', y='Events_by_year', kind='scatter', title='Fireball Event Scatter Plot')
```

```python
#Test train split for supervised training
X_train, X_test, y_train, y_test = train_test_split(df.Year, df.Events_by_year)
```

```python
#Test train split visualizaiton 
plt.scatter(X_train, y_train, label='Training Data', marker = 's', color='r', alpha=.7)
plt.scatter(X_test, y_test, label='Testing Data', color='y', alpha=.7)
plt.legend()
plt.title('Test Train split')
plt.show
```

```python
#Create linear model and train it
LR=LinearRegression()
LR.fit(X_train.values.reshape(-1,1), y_train)
```

```python
#use model to predict on TEST data
prediction = LR.predict(X_test.values.reshape(-1,1))

#plot prediction line againt actual test data
plt.plot(X_test,prediction, label='Linear Regression', color ='b')
plt.scatter(X_test, y_test, label ='Actual Test Data', color ='y', alpha=.7)
plt.legend()
plt.show()
```

```python
#Predict fireball event in 2022
LR.predict(np.array([[2022]]))[0]
```

```python
from sklearn import metrics
metrics.mean_absolute_error(y_test, prediction)
```

```python
metrics.mean_squared_error(y_test, prediction)
```
